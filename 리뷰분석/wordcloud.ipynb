```python
from konlpy.tag import Okt
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import pandas as pd
import psycopg2
```

```python
ex = pd.read_excel('/content/drive/MyDrive/BKMS1_Project/크롤링/BKMS Crawling Data/result_맥북 에어 14인치.xlsx')
```

```python
text = []
for i in range(len(ex)):
  text.append(ex.iloc[i][3])
```

```python
okt = Okt()

text4wc = ''
for sentence in text:
    for noun in okt.nouns(sentence):
        if noun in ['자급', '제', '폰', '에어팟', '세대', '배송', '구매', '제품', '갤럭시', '북', '프로', 
                    '삼성', '노트북','윈도우', '설치', '꼭', '당첨', '생각', '사은', '품','정말', '이벤트', 
                    '아주', '최고', '사용', '다만', '좀', '상품', '하루', '맘', '도', '조금', '무용',
                    '것', '요', '마우스', '용', '엘지', '맥북', '포장', '애플', '주문', '도착']:
            pass
        else:
            text4wc += noun + ' '
```

```python
mask_image = np.array(Image.open('/content/drive/MyDrive/사과.jpg'))
wc = WordCloud(font_path='/content/drive/MyDrive/HMKMRHD.TTF',
               background_color='white',
               max_words=50,
               mask=mask_image,
               max_font_size=200,
               colormap='cool').generate(text4wc)
```

```python
name = ex.iloc[1][1]
wc.to_file('/content/drive/MyDrive/BKMS1_Project/'+name+'.png')
```
