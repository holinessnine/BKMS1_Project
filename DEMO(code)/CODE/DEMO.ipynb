```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
# 색상 언급된 리뷰만 추리는 함수
def color_extract(review):
  color_words = '실버 흰색 화이트 검정 블랙 그레이 스페이스 핑크 레드 골드 색감 색상 컬러 칼라 컬러와 은은한 푸른빛 미드나이트'
  color_words = set(color_words.split(' '))
  result = []
  for idx, row in review.iterrows():
    row = row['review']
    sent_lst = sent_tokenize(row)
    for sent in sent_lst:
      words = sent.split(' ')
      for word in words:  
        if word in color_words:
          result.append(sent)
  return result[:5]
```

```python
# 크기 언급된 리뷰만 추리는 함수
def size_extract(review):
  size_words = '사이즈 사이즈는 커요 크기 작아 작지만 크지만 크기가 크기는'
  size_words = set(size_words.split(' '))
  result = []
  for idx, row in review.iterrows():
    row = row['review']
    sent_lst = sent_tokenize(row)
    for sent in sent_lst:
      words = sent.split(' ')
      for word in words:  
        if word in size_words:
          result.append(sent)
  return result[:5]
```

```python
# 가격 언급된 리뷰만 추리는 함수
def price_extract(review):
  price_words = '가격 가격대 가격은 가격이 비싸 비싼 싼 비싸지만 비싸요 비싸게 싸요 싸지만 싸게 가성비 할인 할인율 저렴한 저렴하다 저렴하게'
  price_words = set(price_words.split(' '))
  result = []
  for idx, row in review.iterrows():
    row = row['review']
    sent_lst = sent_tokenize(row)
    for sent in sent_lst:
      words = sent.split(' ')
      for word in words:  
        if word in price_words:
          result.append(sent)
  return result[:5]
```
